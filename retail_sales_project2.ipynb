# DATA CLEANING

```python
df["Age"].fillna(df["Age"].mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
df.drop_duplicates(inplace=True)
print(df.duplicated().sum())
```

```python
df["Order_Date"]=pd.to_datetime(df["Order_Date"])
df
```

# FEATURE ENGINEERING

```python
df["Total_amount"]=df["Quantity"]*df["Unit_Price"]
df
```

```python
df["Month"]=df["Order_Date"].dt.month_name()
df
```

# EXPLORATORY ANALYSIS

```python
print("Total orders:",df["Order_ID"].count().sum())
```

```python
print("Total_revenue:",df["Total_amount"].count().sum())
```

```python
print("Average_order_value",df["Total_amount"].mean().sum())
```

```python
city_highest_sales=df.groupby("City").agg({"Total_amount":"max"})
city_highest_sales
```

```python
cat_revenue=df.groupby("Product_Category")["Total_amount"].sum()
cat_revenue
cat_revenue.sort_values(ascending=False)
```

```python
avg_age_mean=df.loc[df["Payment_Mode"]=="UPI","Age"].mean()
print(avg_age_mean)
```

# GROUPING AND AGGREGATIONS

```python
df.groupby("City")["Total_amount"].sum()
```

```python
df.groupby("Product_Category").agg({"Quantity":"mean"})
```

```python
avg_amt_bygender=df.groupby("Gender")["Total_amount"].mean()
avg_amt_bygender
```
